```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# https://matplotlib.org/stable/api/pyplot_summary.html

# === STEP 1: Load your data ===
# sample, group, oil_type, oil_conc
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html
df = pd.read_csv("oil_concentration.csv")
#  show the first few rows of data 
df.head()
```

```python
# set colors 
cb_colors = ["#E69F00", "#56B4E9", "#009E73", "#F0E442",
             "#0072B2", "#D55E00", "#CC79A7"]
```

```python
# Set colorblind-friendly palette
sns.set_theme(style="whitegrid", palette=cb_colors)

plt.figure(figsize=(10, 6))

# Create boxplot
sns.boxplot(
    data=df,
    x='group',
    y='oil_conc',
    hue='oil_type'
)

plt.title("Oil Concentration Distribution by Treatment and Oil Type", fontsize=16)
plt.xlabel("Treatment", fontsize=14)
plt.ylabel("Oil Concentration (mg/L)", fontsize=14)
plt.legend(title="Oil Type")
plt.tight_layout()
# === STEP 5: Show or save ===
plt.savefig("figs/oil_concentration_boxplot.png", dpi=300)
plt.show()
```

```python
sns.violinplot(data=df, x='group', y='oil_conc', inner='quartile', palette='colorblind')
```
